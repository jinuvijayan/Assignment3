```python
## how many males and females participated in the test
import numpy as np
import pandas as pd
data=pd.read_csv(r'C:\Users\jinuv\Desktop\Python programs\Assignment3\StudentsPerformance.csv')
data['gender'].value_counts()
```

```python
## Who scores the most on average for math, reading and writing based on Gender and Test preparation course
data.groupby(['gender','test preparation course'])[['math score','reading score','writing score']].mean()
```

```python
## about the scoring variation for math, reading and writing based on Gender and Test preparation course
data.groupby(['gender','test preparation course'])[['math score','reading score','writing score']].var()
```
